# Variable Sharing Between Kernels

This notebook demonstrates how values can be shared across kernel boundaries using `#!share`.
It is the Jupyter `.ipynb` equivalent of `variable-sharing.dib`.

---

## Step 1 — Declare variables in C#

```
var greeting = "Hello";
var target = "Polyglot Notebooks";
Console.WriteLine($"{greeting}, {target}!");
```

## Step 2 — Share the C# variables into JavaScript

The `#!share` magic command copies the named variable from one kernel into another.

```
#!share --from csharp greeting
#!share --from csharp target

console.log(`${greeting}, ${target}! (from JavaScript)`);
```

## Step 3 — Declare a JavaScript variable

```
const jsMessage = "I was born in JavaScript";
```

## Step 4 — Share the JavaScript value back into C#

```
#!share --from javascript jsMessage

Console.WriteLine($"C# received: \"{jsMessage}\"");
```

## Notes

- `#!share --from <kernelName> <variableName>` imports a variable from another kernel into the current one.
- Shared values are serialized (typically as JSON), so complex types must be serializable.
- This feature enables multi-language data pipelines within a single notebook.
