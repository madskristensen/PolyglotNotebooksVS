# Mixed Format Notebook (.ipynb)

This notebook demonstrates multiple languages in a single `.ipynb` file.
Each code cell carries a `polyglot_notebook` metadata entry to declare its kernel.

## C# Cell

```
var message = "Hello from C#";
Console.WriteLine(message);
display(new { Language = "C#", Version = 12, Status = "OK" });
```

## JavaScript Cell

```
const message = 'Hello from JavaScript';
console.log(message);
console.log({ language: 'JavaScript', ecma: 2024, status: 'OK' });
```

## C# — Back to C# after switching kernels

```
// Kernel state is maintained across non-contiguous cells of the same kernel.
// The variable declared in the first C# cell is still in scope here.
Console.WriteLine($"Still in scope: \"{message}\"");
```

## Notes

- The `polyglot_notebook.kernelName` metadata field tells `.NET Interactive` which kernel to use for each cell.
- The `dotnet_interactive.language` field is the legacy equivalent — both are supported.
- Kernel state (variables, imports) is preserved within each kernel across the entire session,
  regardless of the order cells appear in the file.
