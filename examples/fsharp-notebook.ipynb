# F# Notebook

This notebook demonstrates F# features using the `.net-fsharp` kernel.
It tests kernel-name normalization (`.net-fsharp` → `fsharp`) and exercises
let bindings, pattern matching, pipe operators, and list processing.

---

## Let Bindings and Basic Expressions

```
let greeting = "Hello from F#!"
let pi = 3.14159
let circleArea r = pi * r * r

printfn "%s" greeting
printfn "Pi \u2248 %.5f" pi
printfn "Area of circle with r=5: %.5f" (circleArea 5.0)
```

## Lists and Pipe Operators

```
let numbers = [1..10]

let result =
    numbers
    |> List.filter (fun n -> n % 2 = 0)
    |> List.map (fun n -> n * n)

let total = result |> List.sum

printfn "Squares of evens: %A" result
printfn "Sum: %d" total
```

## Pattern Matching

```
let describe n =
    match n with
    | 1 -> "one"
    | 2 -> "two"
    | 3 -> "three"
    | _ -> "something else"

for n in [1; 2; 3; 99] do
    printfn "%d is %s" n (describe n)
```

## Discriminated Unions

```
type Shape =
    | Circle of radius: float
    | Rectangle of width: float * height: float
    | Triangle of baseLen: float * height: float

let area shape =
    match shape with
    | Circle r        -> System.Math.PI * r * r
    | Rectangle(w, h) -> w * h
    | Triangle(b, h)  -> 0.5 * b * h

let shapes = [ Circle 5.0; Rectangle(4.0, 6.0); Triangle(3.0, 8.0) ]

for s in shapes do
    match s with
    | Circle r        -> printfn "Shape: Circle with radius %.1f, area = %.2f" r (area s)
    | Rectangle(w, h) -> printfn "Shape: Rectangle %.1f x %.1f, area = %.2f" w h (area s)
    | Triangle(b, h)  -> printfn "Shape: Triangle with base %.1f and height %.1f, area = %.2f" b h (area s)
```

## Records and Sequence Expressions

```
type Employee = { Name: string; Department: string; Salary: int }

let employees =
    [ { Name = "Alice";  Department = "Engineering"; Salary = 95000 }
      { Name = "Bob";    Department = "Marketing";   Salary = 72000 }
      { Name = "Carol";  Department = "Engineering"; Salary = 105000 } ]

for e in employees do
    printfn "%-7s | %-11s | $%d" e.Name e.Department e.Salary

let avgEngSalary =
    employees
    |> List.filter (fun e -> e.Department = "Engineering")
    |> List.averageBy (fun e -> float e.Salary)

printfn "Average Engineering salary: $%.2f" avgEngSalary
```
